```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data points
x = np.linspace(0, 10, 100)
y_constant = np.ones_like(x)
y_log = np.log2(x + 1)
y_linear = x
y_nlogn = x * np.log2(x + 1)
y_quadratic = x**2
y_cubic = x**3
y_exponential = 2**x

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the functions
plt.plot(x, y_constant, label='O(1)', color='gray')
plt.plot(x, y_log, label='O(log n)', color='green')
plt.plot(x, y_linear, label='O(n)', color='blue')
plt.plot(x, y_nlogn, label='O(n log n)', color='orange')
plt.plot(x, y_quadratic, label='O(n)', color='red')
plt.plot(x, y_cubic, label='O(n)', color='purple')
plt.plot(x, y_exponential, label='O(2)', color='brown')

# Customize the plot
plt.title('Big O Notation - Aravind N 22BRS1099', fontsize=20)
plt.xlabel('Data Input', fontsize=14)
plt.ylabel('Time', fontsize=14)
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.tight_layout()
plt.show()
```
